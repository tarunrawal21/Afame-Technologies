```python
Total_sales = SD['Sales'].sum()
Total_sales
```

**`Sales By Category`**

```python
Sales_per_category = SD.groupby('Category')['Sales'].sum()
plt.figure(figsize=(8,3))
sns.barplot(x=Sales_per_category.index, y=Sales_per_category.values)
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Sales per Category')
```

**sales Trend over the time**

```python
SD['Year'] = pd.to_datetime(SD['Order Date']).dt.year
```

```python
sales_trends_by_year = SD.groupby('Year')['Sales'].sum()
plt.figure(figsize=(8,3))
sns.barplot(x=sales_trends_by_year.index, y=sales_trends_by_year.values)
plt.title('Sales Trends by Year with')
plt.xlabel('Year')
plt.ylabel('Total Sales')
```

**Sales Trend by months Of Each Year**

```python
SD['Month'] = pd.to_datetime(SD['Order Date']).dt.month
```

```python
sales_trends_by_month_year = SD.groupby(['Year', 'Month'])['Sales'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='Sales', hue='Year', data=sales_trends_by_month_year)
plt.title('Sales Trend by Month of Each Year')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(range(1, 13))
plt.legend(title='Year')
plt.show()
```

 **Top 3 best selling Product**

```python
Top_3_best_selling_product = SD.groupby('Product Name')['Sales'].sum().nlargest(3)
plt.figure(figsize=(10,3))
plt.bar(Top_3_best_selling_product.index, Top_3_best_selling_product.values)
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.title('Top 3 best selling Product')
plt.show()
```
