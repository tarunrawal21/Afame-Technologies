```python
colums = list[hr.columns]
colums
```

```python
# colume info
hr.info()
```

```python
# Checking null valeus
print(hr.isnull().sum())
print("No null values found")
```

```python
# Droping unnecessary columns
hr.drop(['Attrition', 'BusinessTravel','DistanceFromHome','Education','EmployeeCount','Over18','StockOptionLevel', 'TrainingTimesLastYear','YearsWithCurrManager'], axis=1, inplace=True)
```

```python
# Types of department
print(hr['Department'].unique())
#Gender retio
print(hr['Gender'].value_counts())
```

```python
hr.columns
```

```python
# Eliminating redundant entries
hr.drop_duplicates(inplace=True)
hr
```

```python
# sanitizing specific columns
hr['JobRole'].unique()
```

```python
# Number Of Employee Doing overtime
hr['OverTime'].value_counts()
```

```python
# number of Employee doing over time in each Job role
hr.groupby('JobRole')['OverTime'].value_counts()
plt.figure(figsize=(12,5))
sns.countplot(x='JobRole', hue='OverTime', data=hr)
plt.xticks(rotation=90)
plt.title('Number Of Employee Doing overtime')
plt.show()
```
